# **1- Import Libraries**

```python
# 1 import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow import keras
from sklearn.preprocessing import LabelEncoder
```

# **2- Data Exploration**

```python
# 2 -Data exploration
df =  pd.read_csv('/content/heart_disease_uci.csv')
df.columns = ['id', 'age', 'sex', 'origin', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal','target']
df.head(10)
df.info
```

```python
df.describe().T
```

# **3- Data preprocessing**


```python
df.isna().sum()
```

```python
df.trestbps.fillna(132.132,inplace=True)

df.chol.fillna(199.130337,inplace = True)
df.fbs.fillna(True, inplace = True)
df.thalach.fillna(137.545665,inplace = True)
df.oldpeak.fillna(0.878788, inplace = True)
df.exang.fillna(False , inplace = True)
df.restecg.fillna('lv hypertrophy', inplace = True)
df = df.drop(['ca','thal'], axis = 1)

print(df)
```

```python
df.slope.describe()
```

```python
df.slope.fillna('flat', inplace = True)
```

```python
df.isna().sum()
```

```python
df.head(10)
```

```python
col = ['sex', 'origin', 'cp' , 'fbs' , 'restecg' , 'slope' , 'slope' , 'exang'  ]

df = pd.get_dummies(df,columns = col, drop_first=True)
```

```python
df.info()
```

```python
df.head()
```

# **4- Split dataframe to train (features) and Test(target)**

```python
x = df.drop('target' , axis = 1)
y = df.target
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y, test_size = 0.2 , random_state = 1)
```

# **5- Standarize and normalize the features**

```python
# import standardscaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# fit and transform x_train
x_train = scaler.fit_transform(x_train)

# transform x_test
x_test = scaler.transform(x_test)
```

# **6- Building the Neural Network**

```python
num_classes = 5
input_size = 21
model = keras.Sequential([
    # the first dense has 120 cells and activation fn is RELu  and input layer has an input size
    keras.layers.Dense(150, activation='relu', input_shape=(input_size,)),

    # adding Dropout layer with 0.2 percent of dropping out
    keras.layers.Dropout(0.2),

    # add onother hidden layer with 30 cells
    keras.layers.Dense(30, activation='relu'),

    # add output layer with cells number equal to no_of_classes and activation function here is "softmax" (multi_class)
    keras.layers.Dense(num_classes, activation='softmax')
])
```

# **7- Compile the model**
set
1- optimizer
2- loss
3- matrics

```python
model.compile( optimizer = 'adam',
                 loss = 'sparse_categorical_crossentropy',
                  metrics = ['accuracy'])
```

# **8- Train the model on the dataset**
using .fit()

```python
model.fit(x_train, y_train, epochs = 15 , batch_size = 28 , validation_data = (x_test , y_test) ,)
```

# **9- Evaluate Model**

```python
test_loss , test_accuracy = model.evaluate(x_test,y_test)
```

```python
print(f"Accuracy = {test_accuracy*100}%")
```
